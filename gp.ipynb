```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
from scipy import stats
import gp
```

```python
x = stats.norm.rvs(8, 1, 10)
y = 13 + 2 * x + stats.norm.rvs(0, 0.7, len(x))
```

```python
x_prime = np.linspace(6, 11, 100)
theta = 0.01

kernel = gp.SquaredExponential(tau = 0.5, l = 0.5)
process = gp.mu(x, y, x_prime, theta, kernel)

for _ in range(5):
    plt.plot(x_prime, process.rvs())

plt.plot(x, y, 'o')
plt.show()
```
